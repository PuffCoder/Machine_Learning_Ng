## 2. Hoeffding Inequation

$$ P(| \hat{\theta} - \theta^{*} | \geq \epsilon ) \leq 2 e^{-2N\epsilon^{2}}$$
* $ \theta $ : true parameters, $\epsilon$ > 0
    * 不等式左侧表示误差/犯错的概率
* 右侧为误差的上界 (Upper Bound)
* 关于采样次数 N 呈现指数级衰减 (exponential decay)

````python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')
eps = 1e-2
plt.plot(2 * np.exp(-2 * np.arange(100000) * eps**2))
plt.xlabel('Sample Size')
plt.ylabel('Upper Bound')
plt.title('Hoeffding’s Inequality Upper Bound')
plt.show()
````

## 3. examples
* PAC: probably Approximate Correct
* 假如约束 $\theta $ 与真实偏差 $\epsilon = 0.1$, 犯错的概率 $ P(| \hat{\theta} - \theta^{*} | \geq 0.1) \leq 0.05 $, How many times shall we thrown?
  * 一个典型的学习理论问题 (learning theory), 在要求的 bound 下， 我们需要多少样本；
* 我们记 $ \delta = 0.05 $
  
$$ P(| \hat{\theta} - \theta^{*} | \geq \epsilon ) \leq 2 e^{-2N\epsilon^{2}}$$


``` Derivation ```
$$ P(| \hat{\theta} - \theta^{*} | \geq \epsilon ) \leq 2 e^{-2N\epsilon^{2}}$$
$$  2 e^{-2N\epsilon^{2}} \leq \delta $$
$$ -2N \epsilon^{2} \leq \ln(\frac{\delta}{2}) $$
$$ -2N \epsilon^{2} \leq \ln{\delta} - \ln{2} $$
$$ 2N \epsilon^{2} \geq \ln{2} - \ln{\delta}   $$
$$ N \geq \frac{\ln{2} - \ln{\delta}} {2\epsilon^{2} }   $$

Substitute $ \delta = 0.05 ; \epsilon = 0.1$ into function    
``` N = 185 ```

````python
(np.log(2) - np.log(0.05)) / (2*0.01)
````
