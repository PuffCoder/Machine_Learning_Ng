# Decision Tree


## Code 

![Q4](./Figures/examGraph/decision_tree/q4.png)

$H$  impurity function  
$Q_{m}$ training data  
$G$ objective function  
$\theta$ feature / split  
$b$ number of children

1. **$H$ - Impurity Function:** 
   - This is a measure used to determine how well a particular feature (or set of features) splits the data. It quantifies the "impurity" or "disorder" in the subsets of data created by the split. Common impurity functions include Gini impurity, entropy, and classification error.

2. **$Q_{m}$ - Training Data:** 
   - In decision tree algorithms, $Q_{m}$ usually refers to the subset of training data at the $m^{th}$ node of the tree. This subset is what the algorithm considers when deciding on the best split to minimize impurity.

3. **$G$ - Objective Function:**
   - This is the function that the decision tree algorithm tries to optimize when choosing the best split. The objective function often involves minimizing the impurity measured by $H$ across the resulting child nodes after a potential split.

4. **$\theta$ - Feature / Split:**
   - Represents the parameters defining how the split is made at a particular node in a decision tree. This typically includes which feature to split on and the threshold value for making the split.

5. **$b$ - Number of Children:**
   - Refers to the number of subgroups or "children" nodes created from a split in a decision tree. In binary trees, $b$ is usually 2, meaning each split results in two child nodes.


````python
import numpy as np
import matplotlib.pyplot as plt

# Generate a range of probabilities from 0 to 1
p = np.linspace(0.01, 0.99, 500)

# Calculate the impurity functions for binary classification
misclassification_error = 1 - np.maximum(p, 1 - p)
gini_index = 2 * p * (1 - p)
entropy = - (p * np.log2(p) + (1 - p) * np.log2(1 - p))

# Plotting the impurity functions
plt.figure(figsize=(10, 6))
plt.plot(p, misclassification_error, label='Misclassification Error', linestyle='--')
plt.plot(p, gini_index, label='Gini Index')
plt.plot(p, entropy, label='Entropy')
plt.title('Impurity Functions for Binary Classification')
plt.xlabel('Probability of one class (p)')
plt.ylabel('Impurity')
plt.legend()
plt.grid(True)
plt.show()
````

In locally weighted linear regression (LWLR), the weights play a critical role in how the model fits the data. The basic idea is to give more importance to the data points that are closer to the point of interest (where the prediction is being made) compared to those that are further away.

Conceptually, these weights are calculated based on the distance of a data point from the point of interest—the closer the data point, the higher the weight. This approach ensures that the regression model is tailored more closely to the vicinity of the target point, allowing for a more flexible fit that can adapt to local variations in the data. This local adaptation makes LWLR particularly useful for datasets where different subsets of data exhibit different behaviors.

a) 
GANs are called "adversarial" because they involve 2 neural networks -- the generator and the discriminator -- competing against each other in a game-like scenario.

The generator's role is to produce data that is as realistic as possible., atempting to fool the discriminator.

The discriminator's role, on he other hand,  is to accurately distingush between real data (from the training set) and fake data produced by hte gneerator. 

This setup creates a dynamic of oppoisition, or adversarial confilict, wheree each network tries to outsmart the ohter, hence the nae "adversarial".


b) 
The learning process of GANs involves alternating training between the 2 networks:
1. Training the Discriminator:
    The discriminator is trained alone. 
    It is shown a batch of real data (e.g., real images) along with a batch of fake data generated by the genertaor. 
    The goal of the discriminator during htis phase is to learn to correctly classify the real data ad real and the fake data as fake.
2. Training the Generator: 
    After the discriminator is trained, the generator is trained next.
    During this phase, hte generator tries to creater new data that will be classified as real by the discriminator.

    This done by adjusting the generator's parammeters to produce outputs that the discriminator is more likely to mistake for real data.

    The generator's success is measured by whether it can fool hte discriminator.
3. Back and Forth:
    There two steps are erpeated in a loop, with each network being updated badsed on its performance againset the other.
    OVer time, the generator gets better at producting realistic data, and the discriminator gets better at telling real from fake. Ideally, this process leads to a point whree the generator produces perfectly realistic data, and the discriminator becomes a perfect judge, equally likely to guess real or fake for hte highly realistic output of the generator.

This adversarial training process is what allows GANs to generate highly realistic outputs, making them particularly useful in applications like image generation, style transfer, and more.
